```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class EmotionCNN(nn.Module):
  def __init__(self):
    super(EmotionCNN,self).__init__()

    self.conv1= nn.conv2(1,32,3,padding =1)
    self.conv2= nn.conv2(32,64,3,padding =1)
    self.conv3= nn.conv2(64,128,3,padding =1)

    self.pool=nn.MaxPool2d(2,2)
    self.dropout=nn.Dropout(.5)

    self.fc1=nn.linear(128*6*6,128)
    self.fc2=nn.linear(128,7)
  def forward(self,x):
    x=self.bool(F.relu(self.conv1(x)))
    x=self.bool(F.relu(self.conv2(x)))
    x=self.bool(F.relu(self.conv3(x)))

    x=x.view(x.size(0),-1)
    x=self.dropout(F.relu(self.fc1(x)))
    x=self.fc2(x)

    return x






```
